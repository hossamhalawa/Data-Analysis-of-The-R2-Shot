```python
import matplotlib.pyplot as plt
import numpy as np
from mplsoccer import Pitch
from scipy.ndimage import gaussian_filter

# Define the pitch dimensions (in meters, standard pitch)
pitch_length = 120
pitch_width = 80

# Define sequences with the last coordinates of each array (scaled to the pitch dimensions)
last_coordinates = [
    {'x': 87 / 100 * pitch_length, 'y': 66 / 100 * pitch_width},  # Mokawlon
    {'x': 83 / 100 * pitch_length, 'y': 66 / 100 * pitch_width},  # Gonna
    {'x': 78 / 100 * pitch_length, 'y': 59 / 100 * pitch_width},  # ItihadShorta
    {'x': 83 / 100 * pitch_length, 'y': 63 / 100 * pitch_width},  # Alhaly2020
    {'x': 87 / 100 * pitch_length, 'y': 68 / 100 * pitch_width},  # Alhaly2021
    {'x': 87 / 100 * pitch_length, 'y': 64 / 100 * pitch_width},  # Haras
    {'x': 77 / 100 * pitch_length, 'y': 62 / 100 * pitch_width},  # Pyramids
    {'x': 87 / 100 * pitch_length, 'y': 66 / 100 * pitch_width},  # ElWaslVsAgman
    {'x': 92 / 100 * pitch_length, 'y': 63 / 100 * pitch_width},  # RaadvsGreece
    {'x': 87 / 100 * pitch_length, 'y': 64 / 100 * pitch_width},  # RaadvsKadesia
]

# Create the football pitch with standard dimensions
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white', pitch_length=pitch_length, pitch_width=pitch_width)
fig, ax = pitch.draw(figsize=(10, 7))

# Extract x and y coordinates for the heatmap
x_coords = [coord['x'] for coord in last_coordinates]
y_coords = [coord['y'] for coord in last_coordinates]

# Create the heatmap with finer bins for accuracy
heatmap, xedges, yedges = np.histogram2d(x_coords, y_coords, bins=[120, 80], range=[[0, pitch_length], [0, pitch_width]])

# Apply Gaussian smoothing to the heatmap with a lower sigma value for accuracy
heatmap = gaussian_filter(heatmap, sigma=2)  # Lower sigma for slight smoothing

# Plot the heatmap on the pitch
pos = ax.imshow(heatmap.T, extent=[0, pitch_length, 0, pitch_width], origin='lower', cmap='hot', alpha=0.7)

# Add a colorbar to show the heat intensity
fig.colorbar(pos, ax=ax)

plt.title("Heatmap of Last Coordinates from Each Sequence", fontsize=16)
plt.show()
```
